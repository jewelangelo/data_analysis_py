```python
import pandas as pd
import os
import shutil
path = os.chdir('C:/Users/jewel.espiritu/OneDrive - AMDATEX LAS PINAS SERVICES INC/Python/py') 
sladf = pd.concat(pd.read_excel('./data/WeeklySLA_Data_2022.xlsx', sheet_name=None), ignore_index=True)
```

```python
SLApivot = pd.pivot_table(sladf, index=['Service','Week'], columns=['TTR Result\n(Excluding Pending Time)'], values=['Ticket No'], aggfunc='count')
Cutoffpivot = pd.pivot_table(sladf, index=['Filing Cut-off','Week'], columns=['Operational Status'], values=['Ticket No'], aggfunc='count')
```

```python
SLApivot = SLApivot.query('Week == 26')
SLApivot.columns = SLApivot.columns.droplevel(0)
SLApivot.columns.name = None
newdf = SLApivot.reset_index()

Cutoffpivot = Cutoffpivot.query('Week == 26')
Cutoffpivot.columns = Cutoffpivot.columns.droplevel(0)
Cutoffpivot.columns.name = None
nnewdf = Cutoffpivot.reset_index()
```

```python
## For Service Count
newdf['Beyond TAT'] = newdf['Beyond TAT'].fillna(0)
newdf['Within TAT'] = newdf['Within TAT'].fillna(0)

newdf['Beyond TAT'] = newdf['Beyond TAT'].astype(str).apply(lambda x: x.replace('.0',''))
newdf['Within TAT'] = newdf['Within TAT'].astype(str).apply(lambda x: x.replace('.0',''))

newdf["Total Count"] = newdf["Beyond TAT"].astype(int) + newdf["Within TAT"].astype(int)

## For Filing Cutoff
nnewdf['Pending'] = nnewdf['Pending'].fillna(0)
nnewdf['On-going'] = nnewdf['On-going'].fillna(0)
nnewdf['Resolved/ Closed'] = nnewdf['Resolved/ Closed'].fillna(0)
nnewdf['Pending'] = nnewdf['Pending'].astype(str).apply(lambda x: x.replace('.0',''))
nnewdf['On-going'] = nnewdf['On-going'].astype(str).apply(lambda x: x.replace('.0',''))
nnewdf['Resolved/ Closed'] = nnewdf['Resolved/ Closed'].astype(str).apply(lambda x: x.replace('.0',''))
```

```python
FilePath = r'C:\repo\SLA_Service.xlsx'
# Create a Pandas Excel writer using XlsxWriter as the engine.
writer = pd.ExcelWriter(FilePath, engine='xlsxwriter')

# Convert the dataframe to an XlsxWriter Excel object.
newdf.to_excel(writer, sheet_name='Service', index = False)
writer.save()

FfilePath = r'C:\repo\SLA_Filing.xlsx'
# Create a Pandas Excel writer using XlsxWriter as the engine.
wwriter = pd.ExcelWriter(FfilePath, engine='xlsxwriter')

# Convert the dataframe to an XlsxWriter Excel object.
nnewdf.to_excel(wwriter, sheet_name='Filing', index = False)
wwriter.save()
```
