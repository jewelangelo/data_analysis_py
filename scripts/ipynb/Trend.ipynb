Load libraries and create df

```python
import pandas as pd
from IPython.display import display
import os
import shutil
path = os.chdir('C:/Users/jewel.espiritu/OneDrive - AMDATEX LAS PINAS SERVICES INC/Python/py')
sladf = pd.concat(pd.read_excel('./data/WeeklySLA_Data_2022.xlsx', sheet_name=None), ignore_index=True)
```

Create pivot for trend 

```python
Trendpivot = pd.pivot_table(sladf, index=['Week'], columns=['TTR Result\n(Excluding Pending Time)'], values=['Ticket No'], aggfunc='count')
```

Remove the unwanted column level of the pivot table

```python
Trendpivot.columns = Trendpivot.columns.droplevel(0)
Trendpivot.columns.name = None
Trenddf = Trendpivot.reset_index()
```

Add a total count of Beyond TAT and Within TAT

```python
Trenddf["Total Count"] = Trenddf["Beyond TAT"].astype(int) + Trenddf["Within TAT"].astype(int)
Trenddf
```

Save the df into an excel file 

```python
FilePath = r'C:\repo\Trend.xlsx'
# Create a Pandas Excel writer using XlsxWriter as the engine.
writer = pd.ExcelWriter(FilePath, engine='xlsxwriter')

# Convert the dataframe to an XlsxWriter Excel object.
Trenddf.to_excel(writer, sheet_name='TREND', index = False)
writer.save()
```
